```python
import pandas as pd

rules = pd.read_csv("association_rules_week3.csv")
rules.head()
```

```python
top_rules = rules.sort_values(by="lift",ascending=False).head(10)

top_rules
```

```python
insights = []

for i, row in top_rules.iterrows():
    antecedent = row["antecedents"]
    consequent = row["consequents"]
    support = row["support"]
    confidence = row["confidence"]
    lift = row["lift"]

    insight = f"If a customer buys {antecedent}, they often also buy {consequent}."
    insights.append(insight)

insights
```

```python
recommendations = []

for i, row in top_rules.iterrows():
    antecedent = row["antecedents"]
    consequent = row["consequents"]

    rec = f"Place {consequent} near {antecedent} in the store OR show it as a combo suggestion online."
    recommendations.append(rec)

recommendations
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,5))
plt.scatter(top_rules["lift"],top_rules["confidence"])
plt.xlabel("lift")
plt.ylabel("confidence")
plt.title("Top rules:lift vs confidence")
plt.grid()
plt.show()
```

```python
summary = pd.DataFrame({
    "Rule": insights,
    "Recommendation": recommendations,
    "Support": top_rules["support"].values,
    "Confidence": top_rules["confidence"].values,
    "Lift": top_rules["lift"].values
})

summary.to_csv("week4_summary.csv", index=False)

summary
```

```python
print("EXECUTIVE SUMMARY\n")

print("Top insights from Apriori Algorithm:\n")
for line in insights:
    print("-",line)

print("\nRecommendations:\n")
for line in recommendations:
    print("-",line)
```
